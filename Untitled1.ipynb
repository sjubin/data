```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
mcdon = pd.read_csv('mcdonalds.csv',index_col='Date',parse_dates=True)
mcdon.head()
```

```python
mcdon['Adj.Volume'].plot(figsize=(12,4)) # 사이즈 늘림 12 =  x 축 , 4 = y 축
```

```python
mcdon['Adj.Close'].plot(xlim=['2007-01-01','2009-01-01'],ylim=([0,50]))
```

```python
mcdon['Adj.Close'].plot(xlim=['2007-01-01','2009-01-01'],ylim=(20,50),ls='--',c='red')
```

```python
idx=mcdon.loc['2007-01-01':'2007-05-01'].index
idx
```

```python
stock= mcdon.loc['2007-01-01':'2007-05-01']['Adj.Close']
stock
```

```python
#plot_date 함수 활용(보기 안 좋음)
fig,ax=plt.subplots()
ax.plot_date(idx,stock)
```

```python
#보기좋게 - 추가해서 한눈에 보기 좋게 함
fig,ax=plt.subplots()
ax.plot_date(idx,stock,'-')
```
